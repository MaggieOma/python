### 2. Append a "world" to the following word.

```python
# Define the initial word
initial_word = "Hello"

# Append "world" to the initial word
result_word = initial_word + " world"

# Print the resulting word
print(result_word)

```

```python
# Define the word
word = "supercalifragilisticexpialidocious"

# Calculate the length of the word
length = len(word)

# Calculate the starting index for the middle six letters
start_index = (length // 2) - 3

# Extract the middle six letters
middle_six_letters = word[start_index:start_index + 6]

print(middle_six_letters)
```
